6. Draw a line graph with X axis (year) and Y axis (electricity consumption of India and Pakistan). Use legend.

```python
# Data for India and Pakistan
india_consumption = [1224.5, 1303.9, 1386.4, 1479.8, 1407.0, 1523.3, 1650.0]
pakistan_consumption = [144.3, 150.5, 159.5, 168.9, 160.9, 172.4, 184.0]

# Create years list
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022]

plt.plot(years, india_consumption, "--ro", label="India", )
plt.plot(years, pakistan_consumption, "-^", label="Pakistan", color="orangered")
plt.legend()
plt.show()
```

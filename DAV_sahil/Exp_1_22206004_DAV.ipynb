```r
library(ggplot2)
library(dplyr)
ggplot(data = mtcars)+
  labs(title = "MTCars Data Plot")
```

```r
#Aesthetic Layer
ggplot(data = mtcars,aes(x = hp, y = mpg, col = disp ))+labs(title = "MTCars Data Plot")
```

```r
# Geometric Layer
ggplot(data = mtcars,aes(x = hp, y = mpg, col = disp ))+ geom_point() + labs(title = "Miles Per Gallon vs Horsepower",x = "Horsepower",y = "Miles Per Gallon")
```

```r
#Adding size
ggplot(data = mtcars,aes(x = hp, y = mpg, size = disp ))+ geom_point() + labs(title = "Miles Per Gallon vs Horsepower",x = "Horsepower",y = "Miles Per Gallon")
```

```r
#Adding shape and color
ggplot(data = mtcars,aes(x = hp, y = mpg, col = factor(cyl), shape = factor(am))) +
  geom_point() +
  labs(title = "Miles Per Gallon vs Horsepower",x = "Horsepower",y = "Miles Per Gallon")
```

```r
#Histogram plot
ggplot(data = mtcars,aes(x = hp))+ geom_histogram(binwidth = 5) + labs(title = "Histogram Of Horsepower",x = "Horsepower",y = "Count")
```

```r
cardio_data <- read.csv("CardioGoodFitness.csv", header = TRUE)
```

```r
#1. Calculate the mean, median of the Fitness column.
mean_fitness <- mean(cardio_data$Fitness)
median_fitness <- median(cardio_data$Fitness)

print(paste("Mean of Fitness:", mean_fitness))
print(paste("Median of Fitness:", median_fitness))
```

```r
#2. Display the minimum and maximum value of Fitness column.
min_fitness <- min(cardio_data$Fitness)
max_fitness <- max(cardio_data$Fitness)


print(paste("Minimum Fitness:", min_fitness))
print(paste("Maximum Fitness:", max_fitness))
```

```r
#3. Display the quartiles and Calculate IQR.
quartiles <- quantile(cardio_data$Fitness, probs = c(0.25, 0.5, 0.75))

print(paste("First quartile (Q1):", quartiles[1]))
print(paste("Second quartile (Q2):", quartiles[2]))
print(paste("Third quartile (Q3):", quartiles[3]))

iqr <- IQR(cardio_data$Fitness)

print(paste("IQR:", iqr))
```

```r
#9. Create a linear graph by using maximum attributes of the dataset CO2.
CO2 <- read.csv("CO2.csv",header = TRUE)
data(CO2)
ggplot(CO2, aes(x = conc, y = uptake, color = Type, shape = Type)) +
  geom_line() +
  geom_point() +
  labs(title = "Linear Graph of Maximum CO2",
       x = "Concentration (uL/L)",
       y = "Uptake Rate (umol/m^2 s)")
```

```r
#10. Create a box plot using ggplot considering CO2 Dataset.
data(CO2)
ggplot(CO2, aes(x = Type, y = uptake, fill = Type)) +
  geom_boxplot() +
  labs(title = "Boxplot of CO2 Uptake by Type",
       x = "Type",
       y = "Uptake Rate (umol/m^2 s)")
```

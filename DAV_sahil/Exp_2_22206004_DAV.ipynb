```r
raw_data <- read.csv("Salary_Data.csv")
```

```r
install.packages('caTools')
library(caTools)
```

```r
set.seed(123)
```

```r
split = sample.split(raw_data$Salary, SplitRatio = 2/3)
```

```r
training_set <- subset(raw_data, split == TRUE)
test_set <- subset(raw_data, split == FALSE)
```

```r
regressor = lm(formula = Salary ~ YearsExperience, data = training_set)
```

```r
summary(regressor)
```

```r
y_pred = predict(regressor, newdata = test_set)
```

```r
library(ggplot2)
```

```r
library(ggplot2)
ggplot() +
  geom_point(data = training_set, aes(x = YearsExperience, y = Salary), color = 'red') +
  geom_line(data = training_set, aes(x = YearsExperience, y = predict(regressor, newdata = training_set)), color = 'blue') +
  ggtitle('Salary vs Experience (Training Set)') +
  xlab('Years of experience') +
  ylab('Salary')
```

```r
library(ggplot2)
ggplot() +
  geom_point(data = test_set, aes(x = YearsExperience, y = Salary), color = 'red') +
  geom_line(data = training_set, aes(x = YearsExperience, y = predict(regressor, newdata = training_set)), color = 'blue') +
  ggtitle('Salary vs Experience (Test Set)') +
  xlab('Years of experience') +
  ylab('Salary')
```
